# Support Vector Machines
## Predict Breast Cancer

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Explore Data

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
cancer.keys()
```

```python
# Load data into DataFrame and view first five columns
df_feat = pd.DataFrame(cancer['data'], columns=cancer.feature_names)
df_feat.head(5)
```

### Split Data

```python
# Train_Test split function
from sklearn.model_selection import train_test_split
```

```python
# Define x-variables
X = df_feat
X
```

```python
# Define y-variable
y = cancer['target']
y
```

```python
# Test size = 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### Build Model

```python
# SVM function
from sklearn.svm import SVC
```

```python
# Instantiate a SVM instance
model = SVC()
```

```python
# Fit model to training data
model.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Set up parameter grid to search for optimail hyperparamters
param_grid = {'C':[0.1, 1, 10, 100, 1000], 'gamma':[1, 0.1, 0.001, 0.0001]}
```

```python
# Instantiate a GridSearchCV instance
grid = GridSearchCV(SVC(), param_grid, verbose=3)
```

```python
# Run CV for find optimal parameters
grid.fit(X_train, y_train)
```

```python
# Optimal parameter values
grid.best_params_
```

```python
# Optimal parameter values
grid.best_estimator_
```

### Make Predictions

```python
# Use predict method to make predictions
grid_predictions = grid.predict(X_test)
grid_predictions
```

### Evaluate Model

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print("Classification Report")
print(classification_report(y_test, grid_predictions))
print("\n")
print("Confusion Matrix")
print(confusion_matrix(y_test, grid_predictions))
```
